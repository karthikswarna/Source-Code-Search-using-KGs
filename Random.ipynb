```python
def ie_preprocess(document):
    sentences = nltk.sent_tokenize(document)
    sentences = [nltk.word_tokenize(sent) for sent in sentences]
    sentences = [nltk.pos_tag(sent) for sent in sentences]
    return sentences

def get_pos(tag):
        if tag.startswith('J'):
            return wordnet.ADJ
        elif tag.startswith('V'):
            return wordnet.VERB
        elif tag.startswith('N'):
            return wordnet.NOUN
        elif tag.startswith('R'):
            return wordnet.ADV
        else:
            return wordnet.NOUN
```

```python
sentences = "The syntactic form of this case: STATEMENT: case <expression>: <statements> RULE: case <expression> -> <expression>/<statement>"


sentences = 'Gets the Object associated with the specified key.'
sentences = "Event sent when the application is asked to open its about window"
sentences = "Thrown to indicate that a program has attempted to access an element of an annotation whose type has changed after the annotation was compiled (or serialized)."
sentences = "A skeletal visitor for annotation values with default behavior appropriate for the RELEASE_6 source version."
sentences = "Hash table and linked list implementation of the Set interface, with predictable iteration order." # good example
sentences = "Provides classes that are fundamental to the design of the Java programming language."
sentences = "get youtube video url id"
sentences = "connect to sql"



sentences = ie_preprocess(sentences)
print(sentences)

# grammar = "NP: {<DT>?<JJ>*<NN>}"
# grammar = "NP: {<DT>?<JJ.*>*<NN.*>+}"
# grammar = r"""
#   NP: {<DT|PP\$>?<JJ>*<IN>?<NN>}   # chunk determiner/possessive, adjectives and noun
#       {<NNP>+}                # chunk sequences of proper nouns
# """

grammar = r"""
  NP: {<DT|PP\$>?<JJ.*>*<NN.*>+}          # Chunk sequences of DT, JJ, NN
  PP: {<IN><NP>}               # Chunk prepositions followed by NP
  VP: {<VB.*><NP|PP|CLAUSE>$} # Chunk verbs and their arguments
  CLAUSE: {<NP><VP>}           # Chunk NP, VP
  """
# grammar = r"""
#   NP: {<DT|PP\$>?<JJ.*>*<NN.*>+}          # Chunk sequences of DT, JJ, NN
#   PP: {<IN><NP>}               # Chunk prepositions followed by NP
#   VP: {<VB.*><NP|PP>$} # Chunk verbs and their arguments
#   """

exp = {}
verb = ''
adverb = ''
to = 0
lemmatizer = WordNetLemmatizer()
stop_words = set(nltk.corpus.stopwords.words("english"))
for sentence in sentences:
    cp = nltk.RegexpParser(grammar)
    result = cp.parse(sentence)
    print(result)
    for index, n in enumerate(result):
        # print(n)
        if str(n[-1]).startswith("VB"):
            if verb == '':
                verb = n[0]
        elif n[-1] == 'WRB':
            if adverb == '':
                adverb = n[0]
                print(adverb)
        elif n[-1] == 'TO':
            to = 1
        if isinstance(n, nltk.tree.Tree):
            newn = ParentedTree.convert(n)
            if n.label() == 'PP':
                print(n)
                req = ''
                for leaf in n.leaves():
                    req += lemmatizer.lemmatize(leaf[0], pos=get_pos(leaf[1])) + ' '
                
                words = req.split(' ')[1:]
                exp[(' '.join([w for w in words if w not in stop_words])).strip()] = ['PP', req.split(' ')[0]]
                # exp[' '.join(req.split(' ')[1:])] = ['PP', req.split(' ')[0]]

            elif (n.label() == 'NP' and newn.parent() != 'PP'):
                print(n)
                req = ''
                for leaf in n.leaves():
                    req += lemmatizer.lemmatize(leaf[0], pos=get_pos(leaf[1])) + ' '
                    
                words = req.split(' ')
                if verb != '':
                    if to == 1:
                        exp[(' '.join([w for w in words if w not in stop_words])).strip()] = ['NP', verb + '_to']
                        # exp[req] = ['NP', verb + '_to']
                        verb = ''
                        if adverb != '':
                            adverb = ''
                    elif adverb != '':
                        exp[(' '.join([w for w in words if w not in stop_words])).strip()] = ['NP', verb + '_' + adverb]
                        # exp[req] = ['NP', verb + '_' + adverb]
                        verb = ''
                        adverb = ''
                    else:
                        exp[(' '.join([w for w in words if w not in stop_words])).strip()] = ['NP', verb]
                        # exp[req] = ['NP', verb]
                        verb = ''
                else:
                    exp[(' '.join([w for w in words if w not in stop_words])).strip()] = ['NP', 'is']
                    # exp[req] = ['NP', 'is']
            
            elif n.label() == 'VP':
                req = ''
                for leaf in n.leaves():
                    req += lemmatizer.lemmatize(leaf[0], pos=get_pos(leaf[1])) + ' '
                print(verb)
                if verb != '':  # can be extended like above for NP.
                    words = req.split(' ')
                    exp[(' '.join([w for w in words if w not in stop_words])).strip()] = ['VP', verb]
                else:
                    words = req.split(' ')[1:]
                    exp[(' '.join([w for w in words if w not in stop_words])).strip()] = ['VP', req.split(' ')[0]]

print(exp)


#     print(nltk.ne_chunk(sentence, binary=True))
#     result.draw()


# First thoughts:
## multiple NP consecutively, might be is_a relationship
## If NP's without relations, just add them as is_a or any default relatiosnhip.
## when,... WRB's can be used as a relationship

# can define relationships as is_?, VB, VBZ(V5), thrown when,
# If sentences start with VBZ, use it. Else use is_a.
## If PP, extraword -> NP (y)
# Interface, Class, Enum, Exception, Error, Annotation Types

# Definite noun phrases, Proper nouns, may be... indefinite nouns
# NP's, VP's, PP's
```
